```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips, hue="day",size="size", palette="YlGnBu")
```

```python
sns.histplot(tips['tip'], kde=True,bins=15) #kde to get line
```

```python
sns.distplot(tips['tip'],kde= True,bins=15)
```

```python
sns.barplot(x="sex",y="tip",data=tips,palette="YlGn")
```

```python
sns.boxplot(x="day",y="tip",data=tips, hue="sex",palette="YlGnBu")
```

```python
sns.boxplot(x="day",y="total_bill",data=tips, hue="sex",palette="YlGnBu")
```

```python
sns.stripplot(x="day",y="tip",data=tips,hue="sex",palette="YlGnBu",dodge=True)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips)
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips, kind='reg')
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips, kind='kde',shade=True,cmap="YlGnBu")
```

```python
sns.jointplot(x="tip",y="total_bill",data=tips, kind='hex',cmap="YlGnBu")
```

```python
sns.pairplot(titanic.select_dtypes(['number']),hue="pclass")
```

```python
# Calculate the correlation matrix only for numeric columns
titanic.select_dtypes(['number']).corr()
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(),annot=True, cmap="YlGnBu") #annot to see the values also
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(),annot=True, cmap="coolwarm") #annot to see the values also
```

```python
sns.heatmap(titanic.select_dtypes(['number']).corr(),annot=True, cmap="icefire") #annot to see the values also
```
