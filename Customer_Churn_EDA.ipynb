```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/WA_Fn-UseC_-Telco-Customer-Churn.csv')
df
```

```python
df.head()
```

```python
#Inspection of Data
df.info()
```

replacing blanks with 0 as tenure is 0 and no total charges are recorded

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ","0")
df["TotalCharges"] = df["TotalCharges"].astype(float)
```

```python
df.info()
```

```python
df.isnull().sum().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
#converted 0 and 1 values of seniors citizen to yes.no to make it easier to understand
def conv(value):
  if value == 1:
    return "yes"
  else:
    return "no"

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)
```

```python
df.head(30)
```

```python
ax = sns.countplot(x = 'Churn',data=df)
ax.bar_label(ax.containers[0])
plt.title("Count of customers by churn")
plt.show()
```

```python
from enum import auto
plt.figure(figsize=(3,4))
gb = df.groupby('Churn').agg({'Churn':"count"})
plt.pie(gb['Churn'],labels = gb.index, autopct = "%1.2f%%")
plt.title("Percentage of Churn Customers",fontsize=10)
plt.show()
```

so 26.54% of customers have churned out according to pie chart so will explore the reason behind it

```python
plt.figure(figsize=(5,5))
ax1 = sns.countplot(x='gender',data=df,hue = 'Churn')
ax1.bar_label(ax1.containers[0])
ax1.bar_label(ax1.containers[1])
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(3,3))
ax2 = sns.countplot(x='SeniorCitizen',data=df)
ax2.bar_label(ax2.containers[0])
plt.title("Count of customers by Senior Citizen")
plt.show()
```

```python
plt.figure(figsize=(3,3))
gpb = df.groupby('SeniorCitizen').agg({'SeniorCitizen':"count"})
plt.pie(gpb['SeniorCitizen'],labels=gpb.index, autopct = "%1.2f%%")
plt.title('Senior Citizen in %')
plt.show()
```

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,bins=72,hue='Churn')
plt.show()
```

people who have used our services stayed a long time but who used for  2 months churned

```python
plt.figure(figsize=(4,4))
ax2 = sns.countplot(x='Contract',data=df,hue='Churn')
ax2.bar_label(ax2.containers[0])
ax2.bar_label(ax2.containers[1])
plt.title("Count of customers by Contract")
plt.show()
```

This says that who have taken month to month contract are likely to churn from those who have one or two yrs of contract
