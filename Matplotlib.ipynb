```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = np.random.random(50)*1000
y_data = np.random.random(50)*1000
```

Scatter Plot

```python
plt.scatter(x_data,y_data,color="red",marker="*",s=100,alpha=0.3)
plt.show()
```

Line Plot

```python
years = [2006 + x for x in range(16)]
weights = [80,83,84,85,86,82,81,79,83,80,82,82,83,81,80,79]
#plt.plot(years,weights,color="purple",lw=5, linestyle="--")
plt.plot(years,weights,"r--",lw=3)
plt.show()
```

Bar chart

```python
x=["C++","C#","Python","Java","C"]
y=[20,50,140,1,45] #votes
```

```python
plt.bar(x,y,color="red", align="edge",width=0.4,edgecolor="green",lw=3) #align means the x value at the edge of bar otherwise middle
for i, v in enumerate(y):
    plt.text(i, v + 0.5, str(v))
plt.show()
```

Histogram

```python
ages = np.random.normal(20,1.5,1000) #mean,sd,no of samples
print(ages)
```

```python
plt.hist(ages,bins=10,cumulative=True)
plt.show()
```

Pie Chart

```python
langs = ["Python","C++","Java","C#","C"]
v = [50,24,14,6,17]
#explodes to highlight
explodes = [0,0,0.3,0,0]
plt.pie(v,labels=langs, explode = explodes,autopct="%.2f%%",startangle=40,pctdistance=1.3)
plt.show()
```

Box Plot

```python
heights = np.random.normal(172, 8, 300)
print(heights)
```

```python
plt.boxplot(heights)
plt.show()
```

```python
f = np.linspace(0,10,25)
s = np.linspace(10,200,25)
t = np.linspace(200,210,25)
fo = np.linspace(210,230,25)
data = np.concatenate((f,s,t,fo))
```

```python
plt.boxplot(data)
plt.show()
```

Plot cutomization

```python
years = [2014,2015,2016,2017,2018,2019,2020,2021]
income=[55,56,62,61,72,72,73,75]
income_ticks = list(range(50,81,2))
plt.plot(years,income)
plt.title("Income Growth(in USD)",fontsize=20,fontname='serif')
plt.xlabel("Years")
plt.ylabel("Income(USD)")
plt.yticks(income_ticks,[f"$ {x}k" for x in income_ticks])
plt.show()
```

```python
sa  =[100,102,99,101,101,100,102]
sb = [90,95,102,104,105,103,109]
sc = [110,115,100,105,100,98,95]
```

```python
votes = [10,2,5,16,22]
people = ["A","B","C","D","E"]
plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

```python
from matplotlib import style
#style.use("ggplot")
style.use("dark_background")
votes = [10,2,5,16,22]
people = ["A","B","C","D","E"]
plt.pie(votes, labels=None)
plt.legend(labels=people)
plt.show()
```

Different charts togther

```python
x1, y1= np.random.random(100), np.random.random(100)
x2,y2 = np.arange(100), np.random.random(100)
style.use("ggplot")
plt.figure(1)
plt.scatter(x1,x2)
plt.figure(2)
plt.plot(x2,y2)
plt.show()
```

3D plots

```python
ax = plt.axes(projection="3d")
x = np.random.random(100)
y = np.random.random(100)
z = np.random.random(100)
ax.scatter(x,y,z)
ax.set_title("3D Plot")
ax.set_xlabel("test")
plt.show()
```
